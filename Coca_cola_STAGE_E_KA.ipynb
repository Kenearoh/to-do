```python
import pandas as pd

# Initialize an empty list
used_links = []

# Append elements to the list
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-23-000024/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-22-000042/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-21-000030/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-20-000046/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-19-000042/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-18-000042/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-17-000032/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-16-000076/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-15-000041/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-14-000036/Financial_Report.xlsx')
links.append('https://s3.amazonaws.com/content.stockpr.com/sec/0000021344-13-000050/Financial_Report.xlsx')


# List to hold each DataFrame
df_list = []

# Loop through each URL
for link in links:
    try:
        # Read the Excel file and append to the list
        df = pd.read_excel(link, sheet_name=2)
        df_list.append(df)
    except Exception as e:
        print(f"Error reading {link}: {e}")

# Concatenate all DataFrames into a single DataFrame
if df_list:
    df_total = pd.concat(df_list, ignore_index=True)
    print(df_total.head())
else:
    print("No DataFrames to concatenate.")
```
