Question 3
Which year had the least correlation with ‘Element Code’?

```python
correlation_2014 = df['Element Code'].corr(df['Y2014'])
correlation_2015 = df['Element Code'].corr(df['Y2015'])
correlation_2016 = df['Element Code'].corr(df['Y2016'])
correlation_2017 = df['Element Code'].corr(df['Y2017'])
```

```python
print('Correlation with Element Code:')
print('2014:', correlation_2014)
print('2015:', correlation_2015)
print('2016:', correlation_2016)
print('2017:', correlation_2017)
min_correlation = min(correlation_2014, correlation_2015, correlation_2016, correlation_2017)
print('Minimum correlation:', min_correlation)
```

This means that the minimum correlation year is 2016

Question
What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python
df.isnull().sum()
```

```python
missing_data =  df['Y2014'].isnull().sum()
total_values =  df.shape[0]
percentage_missing = (missing_data / total_values) * 100

print(f"Total missing values: {missing_data}")
print(f"Percentage of missing values: {percentage_missing:.3f}%")
```

Question
What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

```python
mean_2017 =  df['Y2017'].mean()
std_2017 =  df['Y2017'].std()

print(f"Mean for 2017: {mean_2017:.2f}")
print(f"Standard Deviation for 2017: {std_2017:.2f}")
```
